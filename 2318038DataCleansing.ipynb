# 1. Import library

```python
import pandas as pd
```

# 2. Baca Dataset yang kotor

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/2318038DataCleansing.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

df.head()
```

# 3. Menampilkan struktur variable data

```python
print("Struktur Data Netflix:")
print(df.info())

column_types = df.dtypes
print(column_types)
```

# 4. Data Cleansing – Tangani Missing Values

Algoritma yang digunakan:

1. Kolom director kosong diisi dengan "Unknown".

2. Kolom country kosong diisi dengan modus (nilai terbanyak).

3. Kolom rating kosong dihapus barisnya.

4. Kolom date_added diubah menjadi format datetime dan missing value diganti dengan median.

```python
df["director"].fillna("Unknown", inplace=True)

df["country"].fillna(df["country"].mode()[0], inplace=True)

df.dropna(subset=["rating"], inplace=True)

df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
df["date_added"].fillna(df["date_added"].median(), inplace=True)

df.isnull().sum()
```

# 5. Data Cleansing – Hapus Kolom release_year

```python
df.drop(columns=["release_year"], inplace=True)
df.head()
```

# 6. Data Cleansing – Pisahkan kolom duration

Algoritma:

* Ambil angka (duration_value)

* Ambil jenis (duration_type: "min", "Season", "Seasons")

* Hapus kolom asli duration

```python
df[["duration_value", "duration_type"]] = df["duration"].str.extract(r"(\d+)\s*(\w+)")
df.drop(columns=["duration"], inplace=True)
df.head()
```

# 7. Data Cleansing – Deduplikasi Data

Algoritma:

* Jika ada baris duplikat (berdasarkan title + show_id), simpan hanya baris pertama.

```python
duplicates = df[df.duplicated(subset=["title", "show_id"], keep=False)]
print("Jumlah duplikasi ditemukan:", len(duplicates))

df.drop_duplicates(subset=["title", "show_id"], keep="first", inplace=True)

print("Jumlah duplikasi setelah dibersihkan:", df.duplicated().sum())
```

# 8. Data Cleansing – Normalisasi Nilai Kategori

```python
df["type"] = df["type"].str.upper()

df["rating"] = df["rating"].str.strip()

df[["type", "rating"]].head()
```

# 9. Reset Index

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

# 10. Hasil Akhir

```python
print("\nStruktur Data Setelah Cleansing:")
df.info()

print("\nJumlah Missing Value per Kolom:")
df.isnull().sum()

print("\nJumlah Duplikasi:")
df.duplicated().sum()

print("\nContoh Data Akhir:")
df.head()
```
